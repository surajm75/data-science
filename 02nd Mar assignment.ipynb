Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular data visualization library in Python that provides a wide range of high-quality 2D and 3D plots. It is used to create a variety of static, animated, and interactive plots in Python.

Matplotlib is used to:

Create data visualizations and graphical representations of data
Display data in a clear and meaningful way
Identify patterns, trends, and outliers in data
Communicate data insights and findings effectively
Five plots that can be plotted using the Pyplot module of Matplotlib are:

Line plot: Used to display data points connected by straight line segments.
Scatter plot: Used to display the relationship between two variables in a dataset by plotting them as points on a graph.
Histogram: Used to display the distribution of a numerical variable by dividing the data into intervals called bins.
Bar plot: Used to display the relationship between a categorical variable and a numerical variable.
Pie chart: Used to display the proportions of different categories in a dataset as slices of a circular pie

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

Scatter plot: Used to display the relationship between two variables in a dataset by plotting them as points on a graph.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
#print(x)
#print(y)
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Relation b/w X & Y")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function in Matplotlib's pyplot module is used to create multiple subplots in a single figure. It allows you to divide the figure into a grid of rows and columns, and specify which subplot to create based on its position in the grid.

The subplot() function takes three arguments:

nrows: The number of rows in the grid.
ncols: The number of columns in the grid.
index: The index of the subplot to create, counting from the top left corner of the grid.

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)
axs[0, 0].set_title('Line 1')
axs[0, 1].set_title('Line 2')
axs[1, 0].set_title('Line 3')
axs[1, 1].set_title('Line 4')

plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

Bar plot: Used to display the relationship between a categorical variable and a numerical variable.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
import matplotlib.pyplot as plt
plt.bar(company,profit)
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

box plot : This function creates a box-and-whisker plot of the distribution of each column of the data, with the median represented by a line inside a box, and the whiskers representing the range of the data.

```python
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
import matplotlib.pyplot as plt
plt.boxplot(box1)
```
