Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Sure! Seaborn is a popular Python data visualization library that provides a high-level interface for creating attractive and informative plots. Here are five commonly used plots in Seaborn along with their uses:

Scatter Plot: A scatter plot is a graphical representation of two continuous variables, where data points are represented as dots on a 2D plane. It is useful for visualizing the relationship between two variables and identifying patterns or trends.

Line Plot: A line plot, also known as a line chart, is used to display the trend of a variable over time or across categories. It is particularly useful for visualizing time series data, such as stock prices or temperature changes, and for showing trends or patterns.

Bar Plot: A bar plot, or bar chart, is used to display categorical data with rectangular bars. It can be used to compare the values of different categories, such as comparing the sales of different products or the performance of different teams.

Histogram: A histogram is used to visualize the distribution of a single variable. It displays the frequency or count of data points within different bins or intervals. It is useful for understanding the shape, central tendency, and spread of a variable.

Heatmap: A heatmap is a graphical representation of data in a matrix format, where values are represented using colors. It is particularly useful for visualizing patterns and relationships in large datasets, such as correlation matrices or geographic data.

These are just a few examples of the plots that can be created using Seaborn. Seaborn also provides many other types of plots, such as box plots, violin plots, pair plots, and more, which can be used for various purposes depending on the data and the analysis goals. Seaborn's rich set of visualizations makes it a powerful tool for data exploration and analysis

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
fmri = sns.load_dataset("fmri")
sns.lineplot(fmri, x = "timepoint",y = "signal",hue = 'event')
```

```python
sns.lineplot(fmri, x = "timepoint",y = "signal",hue = 'region')
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
titanic = sns.load_dataset("titanic")
sns.boxplot(titanic,x = 'pclass', y= 'age')
```

```python
sns.boxplot(titanic,x = 'pclass', y= 'fare')
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
diamonds = sns.load_dataset("diamonds")
sns.histplot(diamonds,x='price',hue='cut')
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns
iris = sns.load_dataset("iris")
sns.pairplot(iris,hue = 'species')
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights_df = sns.load_dataset("flights")
flights_pivot = flights_df.pivot("month", "year", "passengers")

sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d", cbar=True)

# plt.title("Monthly Passenger Counts")
# plt.xlabel("Year")
# plt.ylabel("Month")

plt.show()
```
